# Extending MHNs for Metastatic Cancer Progression

This notebook reproduces the figures of section 4.

```python
import plotting
import matplotlib.pyplot as plt
import pandas as pd

metmhn_model_log_theta = pd.read_csv("../run/results/metmhn.csv", index_col=0)
```

```python
fig, ax = plt.subplots(figsize=(3, 3))

plotting_kwargs = {
    "max_br": 1.9,
    "min_br": -0.3,
    "abs_max_ir": 4.3,
}

plotting.plot_theta(
    metmhn_model_log_theta, events=metmhn_model_log_theta.columns.to_list(),
    ax=ax, model="metmhn", **plotting_kwargs)
fig.tight_layout(pad=0)
fig.savefig("results/metmhn.pdf")
```
