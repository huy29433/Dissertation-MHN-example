# Mutual Hazard Networks

This notebook reproduces the figures of section 1.

```python
import mhn
import plotting
import matplotlib.pyplot as plt
import pandas as pd

omhn_model = mhn.model.oMHN.load("../run/results/omhn.csv")
cmhn_model = mhn.model.oMHN.load("../run/results/cmhn.csv")
```

```python
fig1, ax1 = plt.subplots(figsize=(2.2, 2.4))
fig2, ax2 = plt.subplots(figsize=(2.2, 2.4))
ax = [ax1, ax2]

plotting_kwargs = {
    "max_br": 1.9,
    "min_br": -0.3,
    "abs_max_ir": 4.3,
}

plotting.plot_theta(cmhn_model.log_theta, events=omhn_model.events,
                    ax=ax[0], model="cmhn", **plotting_kwargs)
plotting.plot_theta(omhn_model.log_theta, events=omhn_model.events,
                    ax=ax[1], model="omhn", **plotting_kwargs)
ax[0].set_ylim(ax[1].get_ylim()[0], None)
fig1.tight_layout(pad=0)
fig2.tight_layout(pad=0)
fig1.savefig("results/cmhn.pdf")
fig2.savefig("results/omhn.pdf")
```
